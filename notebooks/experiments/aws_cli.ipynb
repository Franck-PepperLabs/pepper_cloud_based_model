# Installation de la commande `aws`

## Infos essentielles avant d'installer

https://aws.amazon.com/fr/cli/

L'interface de la ligne de commande AWS (AWS CLI) est un outil unifié qui permet de gérer vos services AWS. Avec un seul outil à télécharger et à configurer, vous pouvez contrôler plusieurs services AWS à partir de la ligne de commande et les automatiser au moyen de scripts.

[Mise en route](https://docs.aws.amazon.com/fr_fr/cli/latest/userguide/cli-chap-welcome.html)

L'AWS Command Line Interface (AWS CLI) est un outil à code source libre qui vous permet d'interagir avec les services AWS à l'aide des commandes du terminal de ligne de commande.

LaAWS CLI version 2 est la version majeure la plus récenteAWS CLI et prend en charge toutes les fonctions les plus récentes

Pour installer laAWS CLI version 2, reportez-vous à la section [Installation ou mise à jour de la version de l'AWS CLI](https://docs.aws.amazon.com/fr_fr/cli/latest/userguide/getting-started-install.html).

## Instructions pour Windows

Session PowerShell en tant qu'administrateur :
```sh
$ msiexec.exe /i https://awscli.amazonaws.com/AWSCLIV2.msi
```

Nouvelle session PowerShell standard:
```sh
$ aws --version
aws-cli/2.11.19 Python/3.11.3 Windows/10 exe/AMD64 prompt/off
```

## Instructions pour Linux (Ubuntu/WSL)
```sh
$ cd ~
$ curl "https://awscli.amazonaws.com/awscli-exe-linux-x86_64.zip" -o "awscliv2.zip"
$ unzip awscliv2.zip
$ sudo ./aws/install
$ rm awscliv2.zip
$ rm -r aws
$ aws --version
aws-cli/2.11.19 Python/3.11.3 Linux/5.10.16.3-microsoft-standard-WSL2 exe/x86_64.ubuntu.20 prompt/off
```

Il est important de garder vos clés d'accès et secrètes en sécurité. Ces informations sont utilisées pour accéder à vos ressources AWS via l'API ou la ligne de commande, et elles ont le même niveau de privilège que l'utilisateur auquel elles sont associées. Si vous perdez les clés d'accès ou si elles sont compromises, vous pouvez les supprimer et en générer de nouvelles à tout moment.

N'oubliez pas que les clés d'accès et secrètes sont sensibles et doivent être traitées avec précaution. Ne les partagez pas publiquement et ne les stockez pas dans un emplacement non sécurisé.

## [Configuration du AWS CLI](https://docs.aws.amazon.com/fr_fr/cli/latest/userguide/cli-chap-configure.html)

Pour configurer votre commande AWS, vous pouvez exécuter la commande `aws configure` et fournir les informations d'identification de l'utilisateur IAM, ainsi que d'autres paramètres tels que la région par défaut. Vous serez invité à fournir l'Access Key ID, la Secret Access Key, la région par défaut et le format de sortie préféré (par exemple, JSON).

Une fois ces informations configurées, votre commande AWS sera prête à être utilisée avec les autorisations spécifiées pour l'utilisateur IAM.

```sh
$ aws configure
AWS Access Key ID [None]: YOUR_ACCESS_KEY
AWS Secret Access Key [None]: YOUR_SECRET_KEY
Default region name [None]: eu-west-1
Default output format [None]: json
```

```sh
aws s3 help
S3()                                                                      S3()

NAME
       s3 -

DESCRIPTION
       This  section  explains  prominent concepts and notations in the set of
       high-level S3 commands provided.

       If you are looking for the low level S3 commands for  the  CLI,  please
       see the s3api command reference page.

   Path Argument Type
       Whenever using a command, at least one path argument must be specified.
       There are two types of path arguments: LocalPath and S3Uri.

       LocalPath: represents the path of a local file or directory.  It can be
       written as an absolute path or relative path.

       S3Uri: represents the location of a S3 object, prefix, or bucket.  This
       must be written in the form s3://mybucket/mykey where mybucket  is  the
       specified  S3 bucket, mykey is the specified S3 key.  The path argument
       must begin with s3:// in order to denote that the path argument  refers
       to  a  S3  object. Note that prefixes are separated by forward slashes.
       For example, if the S3 object myobject had the prefix myprefix, the  S3
       key would be myprefix/myobject, and if the object was in the bucket my-
       bucket, the S3Uri would be s3://mybucket/myprefix/myobject.
```
